```python
import re
import pandas as pd 
import string
from collections import defaultdict
from datetime import datetime

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
nltk.download('stopwords')

from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python

def lemmatize_sentence(df,col):
      new_col=[]
      punctuation = string.punctuation
      st = set(stopwords.words('english'))
            
      new_name=col+"_modified"
      for metier in df[col]:
            metier = " ".join([word for word in metier.split() if word not in st])
            metier = metier.translate(str.maketrans("", "", punctuation))
            metier = re.sub(r"\d+", "", metier)
            metier=metier.lower()  
            lemmatizer = nltk.WordNetLemmatizer()
            tokens = nltk.word_tokenize(str(metier))
            lemmatized_tokens = " ".join([lemmatizer.lemmatize(token, pos="v") for token in tokens])
            new_col.append(lemmatized_tokens)
      df[new_name]=new_col
      return df
```

```python
def string_to_categorical_job(df,col):
    new_col=[]
    for metier in df[col]:
        if re.search("data", metier) is not None and (re.search("manager", metier) is not None or re.search("managers", metier) is not None or re.search("head", metier) is not None or re.search("managment", metier) is not None or re.search("management", metier) is not None  ):
            new_metier="data manager"
        elif re.search("data", metier) is not None and (re.search("analyst", metier) is not None or re.search("analytics", metier) is not None or re.search("specialist", metier) is not None or re.search("report", metier) is not None)  :
            new_metier="data analyst"
        elif re.search("data", metier) is not None and (re.search("scientist", metier) is not None or re.search("science", metier) is not None) :
            new_metier="data scientist"
        elif re.search("data", metier) is not None and re.search("engineer", metier) is not None:
            new_metier="data engineer"
        elif re.search("data", metier) is not None and (re.search("architect", metier) is not None or re.search("configuration", metier) is not None or re.search("administration", metier) is not None):
            new_metier="data architect"
        elif re.search("data", metier) is not None and re.search("entry", metier) is not None:
            new_metier="data entry"
        elif re.search("business", metier) :
            new_metier="business analyst"
        else:
            new_metier="other data job"
        new_col.append(new_metier)
    df[col]=new_col
    return df
```
